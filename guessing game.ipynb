```python
import random

tries = 1
num = random.randint(1, 100)
guess_list = []

while True:
    guess = int(input(f"Guess a number, try #{tries}: "))
    guess_list.append(guess)
    
    if guess < 1 or guess > 100:
        print("OUT OF BOUNDS")
        continue
    
    if guess == num:
        print(f"You've guessed correctly, and it took you {tries} guesses to get it right!")
        break
    else:
        if tries == 1:
            if abs(guess - num) <= 10:
                print("WARM")
            else:
                print("COLD")
        else:
            prev_guess = guess_list[-2]
            prev_diff = abs(prev_guess - num)
            current_diff = abs(guess - num)
            
            if current_diff == 0:
                print(f"You've guessed correctly, and it took you {tries} guesses to get it right!")
                break
            elif current_diff < prev_diff:
                print("WARMER")
            else:
                print("COLDER")
    tries += 1
```
